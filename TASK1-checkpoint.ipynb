```python
data = pd.DataFrame({
    'Country Name': ['Afghanistan', 'Zimbabwe', 'Brazil', 'Japan', 'Canada', 'Germany', 'India'],
    '2022': [40218234, 16320537, 215353593, 125584838, 38594982, 83155031, 1393409038]
})

# List of countries to filter
countries = ['Afghanistan', 'Zimbabwe', 'Brazil', 'Japan', 'Canada']


filtered_data = data[data['Country Name'].isin(countries)]

# Create bar chart
plt.bar(filtered_data['Country Name'], filtered_data['2022'])
plt.xlabel('Country Name')
plt.ylabel('Population in 2022')
plt.title('Population of selected countries in 2022')
plt.xticks(rotation=45, ha='right')
plt.show()
```
